```python
# Import Library
import pandas as pd
import pickle
from google.colab import files

# Import Data Test
!gdown 1B6JKY7igNrXq25DpQAYPmfKgFMMhFNA2
!gdown 1gPVnhxSUNx1Sx53SxxW60Qnr0Ds4xF7E
test_scaled = pd.read_csv('/content/test_scaled.csv')
test_id = pd.read_csv('/content/test_id.csv')

# Import Model
!gdown 1vO8N_ntkGQ2RFVOnQE4IV8YiYeJTIhWz
```

# Prediksi

```python
# Mapping manual dari hasil encoding ke nilai asli
label_mapping = {
    0: 'Background',
    1: 'Benign',
    2: 'Bruteforce',
    3: 'Bruteforce-XML',
    4: 'Probing',
    5: 'XMRIGCC CryptoMiner'
}

# Load the trained model from the pickle file
model_filename = 'model.pkl'
with open(model_filename, 'rb') as file:
    pipeline = pickle.load(file)

# Lakukan prediksi pada data test
y_test_pred_encoded = pipeline.predict(test_scaled)

# Map hasil prediksi encoded ke label asli
y_test_pred = [label_mapping[label] for label in y_test_pred_encoded]

# Ubah test_id menjadi array dan ratakan menjadi satu dimensi
test_id_array = test_id.values.flatten()

# Ensure test_id is flattened to a one-dimensional array
test_id_flattened = test_id.values.flatten()

# Buat DataFrame untuk submission
submission = pd.DataFrame({
    'id': test_id_flattened,  # Menggunakan variabel test_id yang sudah diratakan
    'traffic': y_test_pred  # Hasil prediksi dalam bentuk label asli
})

# Simpan submission sebagai file CSV
submission_filename = 'submission_ExtraTrees.csv'  # Gantilah nama file sesuai kebutuhan
submission.to_csv(submission_filename, index=False)

print(f"Submission file created: {submission_filename}")

# Download the CSV file
files.download(submission_filename)
```
